```python
#create dim_food_category table
%sql
create table restaurant_booking_project.gold_layer.dim_food_category(
    food_category_id BIGINT GENERATED BY DEFAULT AS IDENTITY primary key not null,
    food_category_name string
)
```

```python
#create dim_food table
%sql
create table restaurant_booking_project.gold_layer.dim_food(
    food_id BIGINT GENERATED BY DEFAULT AS IDENTITY primary key not null,
    food_category_id BIGINT,
    food_name string,
    FOREIGN KEY (food_category_id) REFERENCES restaurant_booking_project.gold_layer.dim_food_category(food_category_id)
)
```

```python
#create fact_booking table
%sql
create table restaurant_booking_project.gold_layer.fact_booking(
    booking_id string primary key not null,
    client_name string,
    table_types_id int,
    booking_time timestamp,
    num_of_individuals int,
    is_birthdate boolean,
    FOREIGN KEY (table_types_id) references restaurant_booking_project.gold_layer.dim_table_types(table_id)
)
```

```python
#create fact_order table
%sql
create table restaurant_booking_project.gold_layer.fact_order(
    booking_id string not null,
    food_category_id long,
    food_id long,
    primary key (booking_id, food_category_id, food_id),
    foreign key (booking_id) references restaurant_booking_project.gold_layer.fact_booking(booking_id),
    FOREIGN KEY (food_category_id) references restaurant_booking_project.gold_layer.dim_food_category(food_category_id),
    FOREIGN KEY (food_id) references restaurant_booking_project.gold_layer.dim_food(food_id)
)
```
